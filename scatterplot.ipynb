```python
import matplotlib.pyplot as plt
```

```python
x=[2,3,5,7,7]
y=[50,60,70,100,90]
plt.tick_params(axis='x',color='red')
color=['r','b','g','m','r']
sizes=[255,440,660,800,500]
plt.scatter(x,y,c=color,marker='*',s=sizes,edgecolor='r',alpha=0.4,label='DataScience')
plt.legend()
plt.title("scatter plot",fontsize=30,color='g')
plt.xlabel("x=students",fontsize=15,loc='right')
plt.ylabel("y=markx",fontsize=15,loc='top')

plt.show()
```
